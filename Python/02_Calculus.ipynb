## Limits

```python
# Using SymPy to calculate limits
from sympy import *
x = symbols('x')
f = 1 / x
result = limit(f, x, oo)     # oo stands for infinity
print(result)
```

```python
# Calculate Euler's number with limits in SymPy
n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)
print(result)
```

```python
print(result.evalf())
```

## Derivatives

```python
# A derivative calculator in Python
def derivative_x(f, x, step_size):
    m = (f(x + step_size) - f(x)) / ((x + step_size) - x)
    return m

def my_function(x):
    return x**2

slope_at_2 = derivative_x(my_function, 2, .00001)

print(slope_at_2)
```

```python
# Calculating a derivative in SymPy
from sympy import *
x = symbols('x')       # declare 'x' to SymPy
f = x**2               # use Python syntax to declare function
dx_f = diff(f)         # calculate the derivative of the function
print(dx_f)            # print the derivative
```

```python
# Calculate the slope at x = 2
print(dx_f.subs(x,2))
```

```python
# Calculating partial derivatives with SymPy
from sympy.plotting import plot3d
x,y = symbols('x y')                 # decalare x and y to SymPy
f = 2*x**3 + 3*y**3                  # use Python syntax to declare function
dx_f = diff(f, x)
dy_f = diff(f, y)
print(dx_f)
print(dy_f)
```

```python
# Plot the function
plot3d(f)
```

```python
# Using limits to calculate a slope
x, s = symbols('x s')           # "x" and step size "s"
f = x**2                        # declare function

# slope between two points with gap "s"
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)     # substitute into rise-over-run formula
slope_2 = slope_f.subs(x, 2)                       # substitute 2 for x

# calculate slope at x = 2
result = limit(slope_2, s, 0)    # infinitely approach step size s to 0
print(result)
```

```python
# Using limits to calculate a derivative
x, s = symbols('x s')           # "x" and step size "s"
f = x**2                        # declare function

# slope between two points with gap "s"
slope_f = (f.subs(x, x + s) - f) / ((x+s) - x)     # substitute into rise-over-run formula

# calculate derivative function
result = limit(slope_f, s, 0)    # infinitely approach step size s to 0
print(result)
```

## The Chain Rule

```python
# Calculating the derivative dz/dx with and without the chain rule
from sympy import *
x, y = symbols('x y')

# derivative for first function
_y = x**2 + 1         # need to underscore y to prevent variable clash
dy_dx = diff(_y)

# derivative for second function
z = y**3 - 2
dz_dy = diff(z)

# Calculate derivative with and without chain rule, substitute y function
dz_dx_chain = (dy_dx * dz_dy).subs(y, _y)
dz_dx_no_chain = diff(z.subs(y, _y))

# Prove chain rule by showing both are equal
print(dz_dx_chain)
print(dz_dx_no_chain)
```

```python
print(dy_dx)
```

```python
print(latex(dz_dx_chain))
```
