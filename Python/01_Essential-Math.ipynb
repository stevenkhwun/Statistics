### `SymPy` library

One of the function of the `SymPy` library is to make plots. To create a plot, we first declare the $x$ variable to `SymPy` using `symbols()` function, and then declare the function and finally plot it.

#### 2D plots

````python
# Import the SymPy library
from sympy import *
````

````python
# Charting a linear function in Python using SymPy
x = symbols('x')    # declare the variable
f = 2*x + 1         # declare the function
plot(f)             # plot it
````

````python
# Charting an exponential functio
f = x**2 + 1
plot(f)
````

#### 3D plots

If the functions have two independent variables ($x$ and $y$) and one independent variable (the output of $f(x,y)$, we need to plot this graph on three dimensions to produce a plane of vlues rather than a line. We need the `plot3d()` function to create this 3D plot.

````python
# Import the plot3d() function
from sympy.plotting import plot3d
````

````python
# Declaring a function with two independent variables in Python
x, y = symbols('x y')
f = 2*x + 3*y
plot3d(f)
````

## Summations

### Summations using `range()` function

Recall that the `range()` function in Python is end exclusive, meaning that if you invoke `range(1,4)` it will iterate the numbers 1, 2, and 3. It excludes the 4 as an upper boundary.

````python
# Performing a summation in Python
summation = sum(2*i for i in range(1,6))
print(summation)
````

````python
# Summation of elements in Python
x = [1, 4, 6, 2]
n = len(x)
summation = sum(10*x[i] for i in range(0,n))
print(summation)
````

### Summations using `SymPy`

A summation operation in `SymPy` is performed using the `Sum()` operator. In the following code, we iterate $i$ from 1 through $n$, multiply each $i$, and sum them. But then we use the `subs()` function to specify $n$ as 5, which will then iterate the sum all $i$ element from 1 throuth 5:

````python
# Summations using SymPy
i,n = symbols('i n')

# iterate each element i from 1 to n, then multiply and sum
summation = Sum(2*i,(i,1,n))

# specify n as 5, iterating the numbers 1 through 5
up_to_5 = summation.subs(n, 5)
print(up_to_5.doit())           # the doit() function is needed to execute the expression
````

## Logarithms

### Using `log()` function in `math` library

````python
# Import the log() function from math library
from math import log
# 2 raised to what power gives me 8?
x = log(8, 2)      # the base argument is specified to 2
print(x)
````

````python
# Further example
# 2 raised to what power gives me 10?
x = log(10, 2)
print(x)
````

When we do not supply a base argument to a `log()` function, the base is Euler's number $e$.

````python
from math import log
x = log(10)
print(x)
````

### Using `log()` function in `SymPy` library

The syntax are similar to the `log()` function in `math` library but the output are formulas unless it is integer.

````python
from sympy import *
x = log(10, 2)
print(x)
````

````python
# When the result is an integer
print(log(8, 2))
````

We can use the `evalf()` function to obtain the exact result when the result is not an integer.

````python
print(x.evalf())
````

## Calculating continuous interest using `exp()` function

Recall that the formula for the continuous interest is $A = Pe^{rt}$.

````python
# Calculating continuous interest in Python
from math import exp
p = 100       # principal, starting amount
r = .20       # interest rate, by year
t = 2.0       # time, number of years
a = p * exp(r*t)
print(a)
````
